```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data= pd.read_csv("driver-data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data=data.drop("id", axis=1)
```

```python
data.info()
```

```python
sns.boxplot("mean_dist_day", data=data)
```

```python
sns.boxplot("mean_over_speed_perc", data=data)
```

```python
from sklearn.cluster import KMeans
```

```python
cls= KMeans(n_clusters=2)
```

```python
cls.fit(data)
```

```python
data['cluster']= cls.labels_
```

```python
sns.lmplot("mean_dist_day","mean_over_speed_perc", data=data, hue= "cluster")
```

```python
cls= KMeans(n_clusters=3)
```

```python
cls.fit(data)
```

```python
data['cluster']= cls.labels_
```

```python
sns.lmplot("mean_dist_day","mean_over_speed_perc", data=data, hue= "cluster")
```

```python
cls= KMeans(n_clusters=4)
```

```python
cls.fit(data)
```

```python
data['cluster']= cls.labels_
```
